```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure as fig
import seaborn as sns
%matplotlib inline
```

```python
dates=['1981-01-01','1981-01-02','1981-01-03','1981-01-04','1981-01-05','1981-01-06','1981-01-07','1981-01-08','1981-01-09','1981-01-10']
min_temperature=[20.7,17.9,18.8,14.6,15.8,15.8,15.8,17.4,21.8,20.8]
max_temperature=[34.7,28.9,31.8,25.6,28.8,21.8,22.8,28.4,30.8,32.0]
print(len(dates))
print(len(max_temperature))
print(len(min_temperature))
```

```python

fig ,axes=plt.subplots(nrows=1,ncols=1,figsize=(15,10))
axes.plot(dates,min_temperature,label='min_temperature')
axes.plot(dates,min_temperature,label='max_temperature')
axes.legend()
```

```python
#write in record
sns.set_style('darkgrid')
fig,ax=plt.subplots(nrows=5,ncols=2)
fig.set_size_inches(18.5,10.5)

df=sns.load_dataset('tips')

sns.barplot(x='sex',y='total_bill',data=df,palette='plasma',estimator=np.mean,ax=ax[0,0]).set_title("Bar Plot")
sns.countplot(x='sex',data=df,ax=ax[0,1]).set_title('Count Plot')
sns.boxplot(x='sex',y='total_bill',data=df,hue='smoker',ax=ax[1,0]).set_title("Box Plot")
sns.violinplot(x='sex',y='total_bill',data=df,hue='sex',ax=ax[1,1]).set_title("Violin Plot")
sns.stripplot(x='sex',y='total_bill',data=df,jitter=True,hue='smoker',dodge=True,ax=ax[2,0]).set_title("Strip Plot")
sns.swarmplot(x='day',y='total_bill',data=df,ax=ax[2,1]).set_title("Swarm Plot")
sns.violinplot(x='day',y='total_bill',data=df,ax=ax[3,0])
sns.swarmplot(x='day',y='total_bill',data=df,color='black',ax=ax[3,0]).set_title("Swarm Plot with violean plot")
sns.kdeplot(x="total_bill",data=df,ax=ax[3,1])
sns.boxenplot(x="day",y="total_bill",color="b",scale="linear",data=df,ax=ax[4,0])

sns.catplot(x="day",y="total_bill",color="g",data=df,ax=ax[4,1],kind='bar')
```

```python
#can call Displot or histplot
sns.set_style('whitegrid')
df=sns.load_dataset('iris')
print(df.head())
```

```python
#write in record
sns.distplot(df['petal_length'],kde=True,color='red',bins=30).set_title('Distibution plot')
```

```python
#write in record
sns.histplot(df['petal_length'],kde=True,color='red',bins=30).set_title('Distibution plot')
```

```python

JointGrid=sns.JointGrid(x='petal_length',y='petal_width',data=df)
JointGrid.plot_joint(sns.scatterplot)
JointGrid.plot_marginals(sns.distplot)
g=sns.jointplot(x='petal_length',y='petal_width',data=df , kind='scatter')
g.fig.suptitle('joint plot')
```

```python
#write in record
g=sns.pairplot(df,hue='species',palette='coolwarm')#palette : color style
g.fig.suptitle("pair plot 1")
g.add_legend()
```

```python
pairgrid=sns.PairGrid(data=df)
pairgrid=pairgrid.map_offdiag(sns.scatterplot)
pairgrid=pairgrid.map_diag(plt.hist)
```

```python
pairgrid=sns.PairGrid(data=df)
pairgri=pairgrid.map_upper(sns.scatterplot)
pairgrid=pairgrid.map_diag(plt.hist)
pairgrid=pairgrid.map_diag(sns.kdeplot)
```

```python
g=sns.PairGrid(df,corner=True)
g.map_lower(sns.scatterplot)
g.map_diag(sns.kdeplot)
```

```python
#write in record
sns.distplot(df['petal_length'],kde=True,color='red',bins=30).set_title('Distibution plot')
sns.rugplot(df['petal_length'])
```
